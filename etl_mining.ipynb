## Dataframe For Dimension and Facts

```python
df_dim_branch = df_inventory[['branch']].copy().drop_duplicates().reset_index(drop=True)
df_dim_branch['branch_id'] = df_dim_branch.index + 1
df_dim_branch = df_dim_branch[['branch_id','branch']]
df_dim_branch = df_dim_branch.rename(columns={'branch':'branch_name'})
df_dim_branch
```

```python
df_dim_part = df_inventory[['part_number','part_name','category']].copy().drop_duplicates()
df_dim_part

# Map parts with alternate parts
ALT_PART_MAPPING = {
    'UC-TS-320':   ('UC-TS-320-KMT',     'Compatible with Komatsu D60'),
    'UC-RL-450':   ('UC-RL-450-SH',      'Compatible with Shantui SD22'),
    'GET-TT-120':  ('GET-TT-120-CAT',    'Compatible with CAT 320 bucket'),
    'TY-27.00R49': ('TY-27.00R49-ALT',   'Alternative compound option')
}
# check if part_name has a correspond altername name, if not value is '-'
df_dim_part['alt_part_number'] = df_dim_part['part_number'].map(
    lambda x: ALT_PART_MAPPING.get(x, ('-', '-'))[0]
)
df_dim_part['notes'] = df_dim_part['part_number'].map(
    lambda x: ALT_PART_MAPPING.get(x, ('-', '-'))[1]
)

df_dim_part['part_id'] = range(1, len(df_dim_part) + 1)
df_dim_part = df_dim_part[[
    'part_id', 'part_number', 'alt_part_number', 'part_name', 'category', 'notes'
]]
df_dim_part
```

```python
df_dim_customer  = df_sales[['customer']].copy().drop_duplicates()
df_dim_customer['customer_id'] = range(1, len(df_dim_customer)+1)
df_dim_customer = df_dim_customer[['customer_id','customer']]
df_dim_customer = df_dim_customer.rename(columns={'customer':'customer_name'})
df_dim_customer
```

```python
# stack the dates from sale_date and transfer_date then drop duplicate to keep date unique.

df_dim_date = pd.concat([df_sales['sale_date'], df_transfers['transfer_date']], ignore_index=True).to_frame('full_date')
df_dim_date = df_dim_date.drop_duplicates()

df_dim_date = df_dim_date.sort_values(by='full_date')
df_dim_date = df_dim_date.reset_index(drop=True)

df_dim_date['time_id'] = pd.to_datetime(df_dim_date['full_date']).dt.strftime('%Y%m%d').astype(int)
df_dim_date = df_dim_date[['time_id','full_date']]

df_dim_date['year'] = df_dim_date['full_date'].dt.year
df_dim_date['month'] = df_dim_date['full_date'].dt.month
df_dim_date['day'] = df_dim_date['full_date'].dt.day
df_dim_date = df_dim_date.rename(columns={"time_id": "date_id"})
df_dim_date
```

```python
# create a lookup dictionary of branch name, part_name, customer name and date to it's id

branch_lookup   = dict(zip(df_dim_branch['branch_name'], df_dim_branch['branch_id']))
part_lookup     = dict(zip(df_dim_part['part_number'], df_dim_part['part_id']))
customer_lookup = dict(zip(df_dim_customer['customer_name'], df_dim_customer['customer_id']))
date_lookup     = dict(zip(df_dim_date['full_date'], df_dim_date['date_id']))

fact_sales = df_sales.copy()
fact_sales['branch_id']     = fact_sales['branch'].map(branch_lookup)
fact_sales['customer_id']   = fact_sales['customer'].map(customer_lookup)
fact_sales['part_id']       = fact_sales['part_number'].map(part_lookup)
fact_sales['sale_date_id'] = fact_sales['sale_date'].map(date_lookup)

fact_sales = fact_sales[[
    'sale_id', 'branch_id', 'customer_id', 'part_id',
    'sale_date_id', 'qty', 'unit_price', 'currency'
]]

fact_transfers = df_transfers.copy()
fact_transfers['from_branch_id'] = fact_transfers['from_branch'].map(branch_lookup)
fact_transfers['to_branch_id']   = fact_transfers['to_branch'].map(branch_lookup)
fact_transfers['part_id']        = fact_transfers['part_number'].map(part_lookup)
fact_transfers['transfer_date_id'] = fact_transfers['transfer_date'].map(date_lookup)
fact_transfers = fact_transfers[[
    'transfer_id', 'from_branch_id', 'to_branch_id', 'part_id',
    'transfer_date_id', 'qty'
]]

fact_inventory = df_inventory.copy()
fact_inventory['branch_id']   = fact_inventory['branch'].map(branch_lookup)
fact_inventory['part_id']     = fact_inventory['part_number'].map(part_lookup)
fact_inventory = fact_inventory[[
    'branch_id', 'part_id',
    'stock_on_hand', 'min_stock', 'uom'
]]
```

```python
# DDL
pgConn = pg.connect(
    dbname = "mitra_db",
    user = "user",
    password = "postgres",
    host = "127.0.0.1",
    port = "5434"
)

curr = pgConn.cursor()

curr.execute(
    """
    DROP TABLE IF EXISTS fact_sales CASCADE;
    DROP TABLE IF EXISTS fact_transfers CASCADE;
    DROP TABLE IF EXISTS fact_inventory CASCADE;
    DROP TABLE IF EXISTS dim_branch CASCADE;
    DROP TABLE IF EXISTS dim_customer CASCADE;
    DROP TABLE IF EXISTS dim_part CASCADE;
    DROP TABLE IF EXISTS dim_date CASCADE;
    
    -- Dimension
    CREATE TABLE dim_branch(
        branch_id INT PRIMARY KEY,
        branch_name VARCHAR(15) NOT NULL UNIQUE
    );

    CREATE TABLE dim_customer(
        customer_id INT PRIMARY KEY,
        customer_name VARCHAR(20) NOT NULL UNIQUE
    );

    CREATE TABLE dim_part(
        part_id INT PRIMARY KEY,
        part_number VARCHAR(20) NOT NULL UNIQUE,
        alt_part_number VARCHAR(20),
        part_name VARCHAR(20),
        category VARCHAR(10),
        notes TEXT
    );

    CREATE TABLE dim_date(
        date_id INT PRIMARY KEY,
        full_date DATE NOT NULL UNIQUE,
        year INT NOT NULL,
        month INT NOT NULL,
        day INT NOT NULL
    );
    -- Facts
    CREATE TABLE fact_sales(
        sale_id INT PRIMARY KEY,
        branch_id INT,
        customer_id INT,
        part_id INT, 
        sale_date_id INT,
        qty INT NOT NULL,
        unit_price NUMERIC(12,4) NOT NULL,
        currency VARCHAR(5) DEFAULT 'USD',
        CONSTRAINT fk_branch FOREIGN KEY (branch_id) REFERENCES dim_branch(branch_id),
        CONSTRAINT fk_customer FOREIGN KEY (customer_id) REFERENCES dim_customer(customer_id),
        CONSTRAINT fk_part FOREIGN KEY (part_id) REFERENCES dim_part(part_id),
        CONSTRAINT fk_date FOREIGN KEY (sale_date_id) REFERENCES dim_date(date_id)
    );

    CREATE TABLE fact_transfers(
        transfer_id INT PRIMARY KEY,
        from_branch_id INT,
        to_branch_id INT,
        part_id INT,
        transfer_date_id INT,
        qty INT NOT NULL,
        CONSTRAINT fk_from_branch FOREIGN KEY (from_branch_id) REFERENCES dim_branch(branch_id),
        CONSTRAINT fk_to_branch FOREIGN KEY (to_branch_id) REFERENCES dim_branch(branch_id),
        CONSTRAINT fk_part FOREIGN KEY (part_id) REFERENCES dim_part(part_id),
        CONSTRAINT fk_date FOREIGN KEY (transfer_date_id) REFERENCES dim_date(date_id)
    );

    CREATE TABLE fact_inventory(
        branch_id INT,
        part_id INT,
        stock_on_hand INT NOT NULL,
        min_stock INT NOT NULL,
        uom VARCHAR(10) DEFAULT 'pcs',
        CONSTRAINT fk_branch FOREIGN KEY (branch_id) REFERENCES dim_branch(branch_id),
        CONSTRAINT fk_part FOREIGN KEY (part_id) REFERENCES dim_part(part_id)
    );
"""
)

pgConn.commit()
pgConn.close()
```

```python
# DML
pgConn = pg.connect(
    dbname = "mitra_db",
    user = "user",
    password = "postgres",
    host = "127.0.0.1",
    port = "5434"
)

curr = pgConn.cursor()
curr.executemany(
    """
    INSERT INTO dim_branch(branch_id, branch_name) 
    VALUES(%s, %s)
""",df_dim_branch.values.tolist()
)
curr.executemany(
    """
    INSERT INTO dim_customer(customer_id, customer_name)
    VALUES(%s, %s)
""", df_dim_customer.values.tolist()
)
curr.executemany(
    """
    INSERT INTO dim_date(date_id, full_date, year, month, day)
    VALUES(%s, %s, %s, %s, %s)
""", df_dim_date.values.tolist()
)
curr.executemany(
    """
    INSERT INTO dim_part(part_id, part_number, alt_part_number, part_name, category, notes)
    VALUES(%s, %s, %s, %s, %s, %s)
""", df_dim_part.values.tolist()
)
curr.executemany(
    """
    INSERT INTO fact_inventory(branch_id, part_id, stock_on_hand, min_stock, uom)
    VALUES(%s, %s, %s, %s, %s)
""", fact_inventory.values.tolist()
)
curr.executemany(
    """
    INSERT INTO fact_sales(sale_id, branch_id, customer_id, part_id, sale_date_id, qty, unit_price, currency)
    VALUES(%s, %s, %s, %s, %s, %s, %s, %s)
""", fact_sales.values.tolist()
)
curr.executemany(
    """
    INSERT INTO fact_transfers(transfer_id, from_branch_id, to_branch_id, part_id, transfer_date_id, qty)
    VALUES(%s, %s, %s, %s, %s, %s)
""", fact_transfers.values.tolist()
)
pgConn.commit()
pgConn.close()
```

# Check Validity

```python
pgConn = pg.connect(
    dbname = "mitra_db",
    user = "user",
    password = "postgres",
    host = "127.0.0.1",
    port = "5434"
)
curr = pgConn.cursor()
curr.execute(
    """
SELECT 
    'dim_branch'     AS table_name, COUNT(*) AS row_count FROM dim_branch
UNION ALL SELECT 'dim_customer',    COUNT(*) FROM dim_customer
UNION ALL SELECT 'dim_part',        COUNT(*) FROM dim_part
UNION ALL SELECT 'dim_date',        COUNT(*) FROM dim_date
UNION ALL SELECT 'fact_sales',      COUNT(*) FROM fact_sales
UNION ALL SELECT 'fact_transfers',  COUNT(*) FROM fact_transfers
UNION ALL SELECT 'fact_inventory',  COUNT(*) FROM fact_inventory;
"""
)
# get the data from postgres to python and turn it into dictionary
result = curr.fetchall()
db_counts = dict(result)

# dictionary expected with dataframe as key and row number as value
expected = {
    'dim_branch': len(df_dim_branch),
    'dim_customer': len(df_dim_customer),
    'dim_date': len(df_dim_date),
    'dim_part': len(df_dim_part),
    'fact_inventory': len(fact_inventory),
    'fact_sales': len(fact_sales),
    'fact_transfers': len(fact_transfers)
}
print("TABLE              | EXPECTED â†’ DB     | STATUS")
print("-" * 50)
all_good = True 
# loop the dataframe/table in the dictionary
for table, expected_count in expected.items():
    # get the row count by checking if there's a key. if not default to zero
    db_count = db_counts.get(table, 0)
    status = "PASS" if expected_count == db_count else "FAIL"
    if status == "FAIL":
        all_good = False
    print(f"{table:18} | {expected_count:>5} â†’ {db_count:>5}   | {status}")

print("-" * 50)
print("OVERALL:", "ALL GOOD!" if all_good else "FIX THE FAILS!")
```
